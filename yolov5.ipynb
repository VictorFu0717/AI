```python
# Images
img = 'yolov5/data/images/bus.jpg'  # or file, Path, PIL, OpenCV, numpy, list

# Inference
results = model(img)

# Results
results.crop()[0]['box'][1] # or .show(), .save(), .crop(), .pandas(), etc.

```
